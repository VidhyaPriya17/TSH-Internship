## Exploratory data analysis of retail

```python
#importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#reading the dataset
data = pd.read_csv('SampleSuperstore.csv')
data.head(5)
```

```python
data.duplicated().sum()
```

```python
data = data.drop_duplicates()
data.shape
```

```python
data = data.drop(labels='Postal Code',axis= 1)
```

## data visualization

```python
plt.bar('Sub-Category','Category', data=data)

plt.title('Category vs Sub-Category')

plt.xlabel('Sub-Category')
plt.ylabel('Category')

plt.xticks(rotation=50)

plt.show()
```

```python
data.hist(bins = 50)
plt.show()
```

```python
plt.figure(figsize=(20,15))
sns.countplot(x=data['State'])

plt.xticks(rotation=50)

plt.title('States')

plt.show()
```
